```python
# ==============================================================================
#                      è›‹ç™½è´¨è¡¨é¢æ–‡ä»¶ç”Ÿæˆè„šæœ¬
#
# ç›®çš„: 
# æœ¬è„šæœ¬æ—¨åœ¨å¸®åŠ©ç”¨æˆ·é€šè¿‡ç®€å•çš„æ­¥éª¤ï¼Œä¸ºæŒ‡å®šçš„è›‹ç™½è´¨ï¼ˆPDBæ ¼å¼ï¼‰å’Œ
# é…ä½“ï¼ˆSDF, MOL2ç­‰æ ¼å¼ï¼‰æ–‡ä»¶ç”Ÿæˆä¸€ä¸ªç²¾ç»†çš„è¡¨é¢æ–‡ä»¶ï¼ˆ`.ply`æ ¼å¼ï¼‰ï¼Œ
# ç”¨äºåç»­çš„è¯ç‰©è®¾è®¡ä¸åˆ†æã€‚
#
# æ“ä½œæµç¨‹:
# 1. ä¿®æ”¹ä¸‹æ–¹ã€æ­¥éª¤ 1ã€‘ä¸­çš„è·¯å¾„ä¸ºæ‚¨è‡ªå·±ç³»ç»Ÿä¸­çš„ç»å¯¹è·¯å¾„ã€‚
# 2. æ¿€æ´»æ­£ç¡®çš„condaç¯å¢ƒ (ä¾‹å¦‚: `conda activate surf_maker`)ã€‚
# 3. åœ¨ç»ˆç«¯ä¸­è¿è¡Œæ­¤è„šæœ¬ (ä¾‹å¦‚: `python generate_surface.py`)ã€‚
# 4. è¿è¡ŒæˆåŠŸåï¼Œæœ€ç»ˆçš„ .ply æ–‡ä»¶å°†ä¿å­˜åœ¨æ‚¨æŒ‡å®šçš„è¾“å‡ºç›®å½•ä¸­ã€‚
#
# ==============================================================================

import os
import sys
import numpy as np
import subprocess
import pymesh
import tempfile
from Bio.PDB import PDBParser, PDBIO, Select 
from Bio.PDB import NeighborSearch, Selection
from rdkit import Chem
from scipy.spatial import distance, KDTree
from IPython.utils import io

# ==============================================================================
#                   æ­¥éª¤ 1: é…ç½®æ‚¨çš„è·¯å¾„
# 
# è¯·åŠ¡å¿…ä¿®æ”¹ä¸‹é¢çš„è·¯å¾„ä¸ºæ‚¨è‡ªå·±ç³»ç»Ÿä¸­çš„ç»å¯¹è·¯å¾„ï¼è¿™æ˜¯å”¯ä¸€éœ€è¦æ‚¨æ‰‹åŠ¨ä¿®æ”¹çš„åœ°æ–¹ã€‚
# ä¸ºäº†è§£å†³ä¸åŒæ“ä½œç³»ç»Ÿç‰ˆæœ¬çš„å…¼å®¹æ€§é—®é¢˜ï¼Œæˆ‘ä»¬åŒæ—¶ä½¿ç”¨ APBS-3.0.0 å’Œ APBS-3.4.1ã€‚
# è¯·ç¡®ä¿ä¸¤è€…éƒ½å·²æ­£ç¡®è§£å‹ã€‚
# ==============================================================================

# --- ä¾èµ–è½¯ä»¶çš„æ ¹ç›®å½•è·¯å¾„ ---
# APBS 3.0.0 æ–‡ä»¶å¤¹çš„ç»å¯¹è·¯å¾„
apbs_3_0_0_dir = "/home/cenking/VsCode/Delete/APBS-3.0.0.Linux"

# APBS 3.4.1 æ–‡ä»¶å¤¹çš„ç»å¯¹è·¯å¾„ (ç”¨äºè§£å†³å…¼å®¹æ€§é—®é¢˜)
apbs_3_4_1_dir = "/home/cenking/VsCode/Delete/APBS-3.4.1.Linux" # å‡è®¾æ‚¨å·²ä¸‹è½½å¹¶è§£å‹

# pdb2pqr æ–‡ä»¶å¤¹çš„ç»å¯¹è·¯å¾„
pdb2pqr_dir = "/home/cenking/VsCode/Delete/pdb2pqr-linux-bin64-2.1.1"

# æ‚¨å½“å‰ conda ç¯å¢ƒçš„ç»å¯¹è·¯å¾„ (å¯é€‰ï¼Œä½†å»ºè®®å¡«å†™ä»¥ç¡®ä¿åº“æ–‡ä»¶èƒ½è¢«æ‰¾åˆ°)
conda_env_path = "/home/cenking/miniconda3/envs/surf_maker"

# --- è¾“å…¥æ–‡ä»¶çš„ç»å¯¹è·¯å¾„ ---
# è›‹ç™½è´¨ PDB æ–‡ä»¶çš„ç»å¯¹è·¯å¾„
protein_pdb_path = "/home/cenking/VsCode/Delete/data/workdir/7wmm/7WMM_Macro.pdb"

# é…ä½“æ–‡ä»¶çš„ç»å¯¹è·¯å¾„ (ä¾‹å¦‚ .sdf, .mol2)
ligand_file_path = "/home/cenking/VsCode/Delete/data/workdir/7wmm/7WMM_Ligand1_1_Centered1.sdf"

# --- è¾“å‡ºç›®å½•çš„ç»å¯¹è·¯å¾„ ---
# æ‚¨å¸Œæœ›ä¿å­˜æœ€ç»ˆ .ply æ–‡ä»¶çš„ç›®å½•
output_directory = "/home/cenking/VsCode/Delete/data/workdir/7wmm"


# ==============================================================================
#                   æ­¥éª¤ 2: ç¯å¢ƒè®¾ç½®ä¸è·¯å¾„æ£€æŸ¥
# 
# æ­¤éƒ¨åˆ†å°†è‡ªåŠ¨é…ç½®å’Œæ£€æŸ¥è·¯å¾„ã€‚å¦‚æœä»»ä½•è·¯å¾„ä¸æ­£ç¡®ï¼Œç¨‹åºå°†æŠ¥é”™å¹¶åœæ­¢ã€‚
# ==============================================================================
def setup_and_check_paths():
    print("--- å¼€å§‹é…ç½®å’Œæ£€æŸ¥ç¯å¢ƒ... ---")

    # 1. ç»„åˆå¹¶å®šä¹‰å„ä¸ªå·¥å…·çš„äºŒè¿›åˆ¶æ–‡ä»¶è·¯å¾„
    global msms_bin, apbs_bin, pdb2pqr_bin, multivalue_bin
    msms_bin = os.path.join(apbs_3_0_0_dir, "bin", "msms")
    apbs_bin = os.path.join(apbs_3_4_1_dir, "bin", "apbs")
    pdb2pqr_bin = os.path.join(pdb2pqr_dir, "pdb2pqr")
    multivalue_bin = os.path.join(apbs_3_0_0_dir, "share", "apbs", "tools", "bin", "multivalue")

    # 2. è®¾ç½® LD_LIBRARY_PATH ç¯å¢ƒå˜é‡
    apbs_3_0_0_lib = os.path.join(apbs_3_0_0_dir, "lib")
    apbs_3_4_1_lib = os.path.join(apbs_3_4_1_dir, "lib")
    conda_env_lib = os.path.join(conda_env_path, "lib")
    os.environ["LD_LIBRARY_PATH"] = f"{apbs_3_0_0_lib}:{apbs_3_4_1_lib}:{conda_env_lib}"
    print(f"è®¾ç½® LD_LIBRARY_PATH ä¸º: {os.environ['LD_LIBRARY_PATH']}")

    # 3. æ£€æŸ¥æ‰€æœ‰è·¯å¾„å’Œæ–‡ä»¶æ˜¯å¦å­˜åœ¨
    paths_to_check = {
        "APBS 3.0.0 ç›®å½•": apbs_3_0_0_dir,
        "APBS 3.4.1 ç›®å½•": apbs_3_4_1_dir,
        "PDB2PQR ç›®å½•": pdb2pqr_dir,
        "è›‹ç™½è´¨æ–‡ä»¶": protein_pdb_path,
        "é…ä½“æ–‡ä»¶": ligand_file_path,
        "è¾“å‡ºç›®å½•": output_directory,
        "MSMSç¨‹åº": msms_bin,
        "APBSç¨‹åº": apbs_bin,
        "PDB2PQRç¨‹åº": pdb2pqr_bin,
        "Multivalueç¨‹åº": multivalue_bin
    }

    all_paths_ok = True
    for name, path in paths_to_check.items():
        if not os.path.exists(path):
            print(f"âŒ é”™è¯¯: {name} æœªåœ¨ä»¥ä¸‹è·¯å¾„æ‰¾åˆ°: {path}")
            all_paths_ok = False
        else:
            print(f"âœ… {name} è·¯å¾„æ­£ç¡®.")

    # ç¡®ä¿è¾“å‡ºç›®å½•å­˜åœ¨
    if os.path.exists(output_directory):
        if not os.path.isdir(output_directory):
            print(f"âŒ é”™è¯¯: æŒ‡å®šçš„è¾“å‡ºè·¯å¾„ {output_directory} æ˜¯ä¸€ä¸ªæ–‡ä»¶ï¼Œè€Œä¸æ˜¯ä¸€ä¸ªç›®å½•ï¼")
            all_paths_ok = False
    else:
        print(f"è¾“å‡ºç›®å½• {output_directory} ä¸å­˜åœ¨ï¼Œå°†è‡ªåŠ¨åˆ›å»º...")
        os.makedirs(output_directory)
        print(f"âœ… å·²åˆ›å»ºè¾“å‡ºç›®å½•: {output_directory}")

    if not all_paths_ok:
        raise FileNotFoundError("éƒ¨åˆ†å…³é”®æ–‡ä»¶æˆ–ç›®å½•æœªæ‰¾åˆ°ï¼Œè¯·è¿”å›è„šæœ¬é¡¶éƒ¨æ£€æŸ¥å¹¶ä¿®æ­£è·¯å¾„ï¼")
    else:
        print("\nğŸ‰ ç¯å¢ƒé…ç½®å’Œè·¯å¾„æ£€æŸ¥å…¨éƒ¨é€šè¿‡ï¼")
    
    # 4. åŠ¨æ€æ·»åŠ è¾…åŠ©è„šæœ¬çš„è·¯å¾„
    # è¿™ä½¿å¾—è„šæœ¬å¯ä»¥æ‰¾åˆ° computeMSMS ç­‰æ¨¡å—
    script_dir = os.path.dirname(os.path.abspath(__file__))
    utils_path = os.path.join(script_dir, '..', 'utils', 'masif')
    if not os.path.isdir(utils_path):
         raise FileNotFoundError(f"è¾…åŠ©è„šæœ¬ç›®å½• '{utils_path}' ä¸å­˜åœ¨ã€‚è¯·ç¡®ä¿æ­¤è„šæœ¬ä½äºæ­£ç¡®çš„ 'data' æ–‡ä»¶å¤¹ä¸‹ã€‚")
    sys.path.append(utils_path)
    print(f"âœ… å·²æˆåŠŸæ·»åŠ è¾…åŠ©è„šæœ¬è·¯å¾„: {utils_path}")
    print("--- ç¯å¢ƒæ£€æŸ¥å®Œæ¯• ---")


# ==============================================================================
#                       æ­¥éª¤ 3: ä¸»ç¨‹åº
# 
# è¿™æ˜¯æ ¸å¿ƒè®¡ç®—æ­¥éª¤ã€‚å®ƒå°†è°ƒç”¨ä¸€ç³»åˆ—å·¥å…·æ¥è®¡ç®—å¹¶ç”Ÿæˆè›‹ç™½è´¨è¡¨é¢ã€‚
# ==============================================================================
def main():
    # é¦–å…ˆï¼Œè®¾ç½®å¹¶æ£€æŸ¥æ‰€æœ‰è·¯å¾„
    setup_and_check_paths()
    
    # åŠ¨æ€å¯¼å…¥è¾…åŠ©æ¨¡å—
    from compute_normal import compute_normal
    from computeAPBS import computeAPBS
    from computeCharges import computeCharges, assignChargesToNewMesh
    from computeHydrophobicity import computeHydrophobicity
    from computeMSMS import computeMSMS
    from fixmesh import fix_mesh
    from save_ply import save_ply

    print("\n--- å¼€å§‹æ‰§è¡Œè¡¨é¢ç”Ÿæˆæµç¨‹ ---")

    # --- å‚æ•°è®¾ç½® (é€šå¸¸æ— éœ€ä¿®æ”¹) ---
    dist_threshold=8.0
    use_hbond=True
    use_hphob=True
    use_apbs=True
    mesh_res=1.0
    epsilon=1.0e-6
    feature_interpolation=True

    # ä½¿ç”¨ä¸´æ—¶ç›®å½•æ¥å­˜æ”¾ä¸­é—´æ–‡ä»¶ï¼Œè¿è¡Œç»“æŸåä¼šè‡ªåŠ¨æ¸…ç†
    with tempfile.TemporaryDirectory() as workdir:
        print(f"åˆ›å»ºä¸´æ—¶å·¥ä½œç›®å½•: {workdir}")
        protname = os.path.basename(protein_pdb_path).replace(".pdb","")

        print("1. æ­£åœ¨è¯»å–é…ä½“æ–‡ä»¶...")
        suffix = ligand_file_path.split('.')[-1]
        if suffix == 'mol':
            mol = Chem.MolFromMolFile(ligand_file_path)
        elif suffix == 'mol2':
            mol = Chem.MolFromMol2File(ligand_file_path)
        elif suffix == 'sdf':
            suppl = Chem.SDMolSupplier(ligand_file_path, sanitize=False)
            mols = [mol for mol in suppl if mol]    
            mol = mols[0]
        atomCoords = mol.GetConformers()[0].GetPositions()

        print("2. æ­£åœ¨ç­›é€‰å£è¢‹æ®‹åŸº...")
        parser = PDBParser(QUIET=True)
        structures = parser.get_structure('target', protein_pdb_path)
        structure = structures[0]
        atoms  = Selection.unfold_entities(structure, 'A')
        ns = NeighborSearch(atoms)
        close_residues= []
        for a in atomCoords:  
            close_residues.extend(ns.search(a, dist_threshold+5, level='R'))
        close_residues = Selection.uniqueify(close_residues)

        class SelectNeighbors(Select):
            def accept_residue(self, residue):
                if residue in close_residues:
                    if all(a in [i.get_name() for i in residue.get_unpacked_list()] for a in ['N', 'CA', 'C', 'O']) or residue.resname=='HOH':
                        return True
                    else: return False
                else: return False
        
        pocket_pdb_path = os.path.join(workdir, f"{protname}_pocket_{dist_threshold+5}.pdb")
        pdbio = PDBIO()
        pdbio.set_structure(structure)
        pdbio.save(pocket_pdb_path, SelectNeighbors())
        print(f"å·²å°†å£è¢‹åŒºåŸŸä¿å­˜åˆ°ä¸´æ—¶æ–‡ä»¶: {pocket_pdb_path}")

        print("3. æ­£åœ¨è®¡ç®—åˆ†å­è¡¨é¢ (MSMS)... æ­¤æ­¥éª¤å¯èƒ½éœ€è¦ä¸€äº›æ—¶é—´ã€‚")
        structures = parser.get_structure('target', pocket_pdb_path)
        structure = structures[0]
        atoms = Selection.unfold_entities(structure, 'A')
        dist = [distance.euclidean(atomCoords.mean(axis=0), a.get_coord()) for a in atoms]
        atom_idx = np.argmin(dist)
        vertices1, faces1, normals1, names1, areas1 = computeMSMS(pocket_pdb_path,  
                                                                    protonate=True, 
                                                                    one_cavity=atom_idx, 
                                                                    msms_bin=msms_bin,
                                                                    workdir=workdir)

        print("4. æ­£åœ¨è®¡ç®—è¡¨é¢ç‰¹å¾å’Œä¿®å¤ç½‘æ ¼...")
        kdt = KDTree(atomCoords)
        d, r = kdt.query(vertices1)
        iface_v = np.where(d <= dist_threshold)[0]
        faces_to_keep = [idx for idx, face in enumerate(faces1) if all(v in iface_v  for v in face)] 
        
        vertex_hbond = None
        if use_hbond:
            vertex_hbond = computeCharges(protein_pdb_path.replace(".pdb",""), vertices1, names1)
        
        vertex_hphobicity = None
        if use_hphob:
            vertex_hphobicity = computeHydrophobicity(names1) 

        mesh = pymesh.form_mesh(vertices1, np.array(faces1))
        mesh = pymesh.submesh(mesh, faces_to_keep, 0)
        with io.capture_output() as captured:
            regular_mesh = fix_mesh(mesh, mesh_res)

        regular_mesh, info = pymesh.remove_degenerated_triangles(regular_mesh)
        vertex_normal = compute_normal(regular_mesh.vertices, regular_mesh.faces, eps=epsilon)

        if use_hbond and vertex_hbond is not None:
            vertex_hbond = assignChargesToNewMesh(regular_mesh.vertices, vertices1, vertex_hbond, feature_interpolation)
        if use_hphob and vertex_hphobicity is not None:
            vertex_hphobicity = assignChargesToNewMesh(regular_mesh.vertices, vertices1, vertex_hphobicity, feature_interpolation)

        vertex_charges = None
        if use_apbs:
            print("5. æ­£åœ¨è®¡ç®—é™ç”µåŠ¿ (APBS)... æ­¤æ­¥éª¤å¯èƒ½éœ€è¦ä¸€äº›æ—¶é—´ã€‚")
            vertex_charges = computeAPBS(regular_mesh.vertices, pocket_pdb_path, 
                                        apbs_bin, pdb2pqr_bin, multivalue_bin, workdir)
        
        print("6. æ­£åœ¨è®¡ç®—å‡ ä½•ç‰¹å¾...")
        regular_mesh.add_attribute("vertex_mean_curvature")
        H = regular_mesh.get_attribute("vertex_mean_curvature")
        regular_mesh.add_attribute("vertex_gaussian_curvature")
        K = regular_mesh.get_attribute("vertex_gaussian_curvature")
        elem = np.square(H) - K
        elem[elem<0] = 1e-8
        k1 = H + np.sqrt(elem)
        k2 = H - np.sqrt(elem)
        si = (k1+k2)/(k1-k2)
        si = np.arctan(si)*(2/np.pi)

        output_ply_path = os.path.join(output_directory, f"{protname}_pocket_{dist_threshold}.ply")
        print(f"7. æ­£åœ¨ä¿å­˜æœ€ç»ˆç»“æœåˆ°: {output_ply_path}")
        save_ply(output_ply_path, regular_mesh.vertices,
                regular_mesh.faces, normals=vertex_normal, charges=vertex_charges,
                normalize_charges=True, hbond=vertex_hbond, hphob=vertex_hphobicity,
                si=si)

    print("\n" + "="*50)
    print("ğŸ‰ğŸ‰ğŸ‰ æ­å–œï¼å¤„ç†å®Œæˆï¼ ğŸ‰ğŸ‰ğŸ‰")
    print(f"æœ€ç»ˆçš„è¡¨é¢æ–‡ä»¶å·²ä¿å­˜è‡³: \n{output_ply_path}")
    print("="*50)


# ==============================================================================
# å½“è„šæœ¬è¢«ç›´æ¥æ‰§è¡Œæ—¶ï¼Œè¿è¡Œmain()å‡½æ•°
# ==============================================================================
if __name__ == "__main__":
    main()
```
